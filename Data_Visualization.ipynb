## Pareto and SLudge

```python
file_paths = [r'C:\Users\Maryam\Desktop\Python\Thesis\classes\Values_class_no_5_area.csv',
              r'C:\Users\Maryam\Desktop\Python\Thesis\classes\Values_class_no_10_area.csv',
              r'C:\Users\Maryam\Desktop\Python\Thesis\classes\Values_class_no_15_area.csv',
              r'C:\Users\Maryam\Desktop\Python\Thesis\classes\Values_class_no_20_area.csv']

data=pd.DataFrame()
for file_path in file_paths:
    # read
    df=pd.read_csv(file_path)
    # remove space from columns
    df.columns=df.columns.str.strip()
    # combines all files together
    data=pd.concat([data,df],ignore_index=True)
    
# remove duplications from configurations
unique_nc=data['nc'].unique() 

unique_nc = unique_nc.astype(int) 

# remove duplications from configurations
unique_sludge=data['Sludge'].unique()
unique_sludge=unique_sludge.astype(int) 

marker_styles=['o', 's', 'D', '^', 'v', '<', '>', 'p', 'H', 'X']

# Convert unique_configurations to a Python list
unique_nc_list = unique_nc.tolist()
unique_sludge_list=unique_sludge.tolist() 

#configurations_to_markers=dict(zip(unique_configurations_list, marker_styles))

#colourmap=plt.cm.get_cmap('viridis', len(unique_configurations_list)*len(unique_sludge_list))
colors = plt.cm.viridis(np.linspace(0, 1, len(unique_nc) * len(unique_sludge)))
plt.figure(figsize=(12,8))


for i, unique_nc in enumerate(unique_nc_list):
    for j, unique_sludge in enumerate(unique_sludge_list):
        config_data = data[(data['nc'] == unique_nc) & (data['Sludge']==unique_sludge)]
        marker_style=marker_styles[i % len(marker_styles)]
        color=colors[i*len(unique_sludge_list)+j]
        plt.scatter(config_data['TAC'], config_data['CO2e'], color=color,label=f'{unique_nc} - {unique_sludge}',
                marker=marker_style)
# add labels 
#plt.title('Parallel Coordinates Visualization')
# Function to format ticks with commas
def format_func(value, tick_number):
    return "{:,}".format(int(value))

# Apply the formatter to the x-axis
plt.gca().xaxis.set_major_formatter(ticker.FuncFormatter(format_func)) 
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(format_func)) 
       
plt.xlabel('Total Annualized Cost ($/yr)')
plt.ylabel('Total CO2 Emission (KgCO2e/yr)')
plt.legend()
# show plot 
plt.show()        
```

## Re vs Cost of CPI

```python


# Constants
h_cpi = 0.0127
rhoC = 980
qfeed = 250
fConversion = 0.000274
MuW = 0.00054
oil_skimmer = 459
As = 49.35079035290751

# Arrays for w_cpi and l_cpi
w_cpi = [i for i in np.arange(1.8, 6, 0.1)]
l_cpi = [i for i in np.arange(0.5, 4.5, 0.1)]

# Initialize lists for data
x_data, y_data, z_data, w_data = [], [], [], []

for w in w_cpi:
    for l in l_cpi:
        # Calculate variables
        diameter = (2 * h_cpi * w) / (h_cpi + w)
        Area = l * w
        velocity = qfeed * fConversion / Area
        Re = rhoC * velocity * diameter / MuW
        L_tank = (12 / 5) * l
        Ap = l * w
        n_cpi = round(As / Ap)
        H_cpi = (n_cpi - 1) * h_cpi
        w_tank = w
        H_tank = H_cpi + 0.45
        vol_cpi = H_cpi * l * w
        vol_tank = H_tank * L_tank * w_tank
        Ce_total = 5800 + 1600 * vol_tank ** 0.7 + 500 * vol_cpi + oil_skimmer
        
        # Append data to lists
        x_data.append(w)
        y_data.append(l)
        z_data.append(Re)
        w_data.append(Ce_total)

# Plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
scatter = ax.scatter(x_data, y_data, z_data, c=w_data, cmap='viridis')
plt.colorbar(scatter, label='Cost ($/yr)')
ax.set_xlabel('w_cpi (m)')
ax.set_ylabel('l_cpi (m)')
ax.set_zlabel('Re')
plt.show()
        

```
